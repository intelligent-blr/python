```python
# Первый вариант калькулятора
a, b = int(input('Введите первое число:')), int(input('Введите второе число:'))
print(a + b)
print(a - b)
print(a * b)
try:                                         # try - что делаем 
    print(a / b)
except ZeroDivisionError:                    # except - исключение, ZeroDivisionError - тип ошибки для исключения
    print("Ошибка: Нельзя делить на ноль.")
try:
    print(a % b)                             # взятие остатка от деления
except ZeroDivisionError: 
    print("Ошибка: Нельзя делить на ноль.")
try:
    print(a // b)                            # целочисленное деление
except ZeroDivisionError: 
    print("Ошибка: Нельзя делить на ноль.")
print(a ** b)
```

```python
# Второй вариант калькулятора
def calculator(a, b, operation):
    if operation == '+':                     # if - если, elif - или если, else - то
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '*':
        return a * b
    elif operation == '/' and b !=0:
        return a / b,
    elif operation == '*':
        return a % b
    elif operation == '/' and b !=0:
        return a // b,   
    elif operation == '**':
        return a ** b,
    else: 
        return 'Wrong operation'
calculator(8,2, '**')
```

```python
r = float(input('Введите радиус окружности: '))
pi = 3.14159
print(f'Длина окружности: {2 * pi * r}')
print(f'Площадь окружности: {pi * r ** 2}')
```

```python
# map - используется для применения функции к каждому элементу итерируемого объекта
# .slpit() - преобразует строку в list по разделителю - по умолчанию это пробел
x1, y1 = map(int, input('Введите координаты первой точки (x1, y1): ').split(',')) 
x2, y2 = map(int, input('Введите координаты второй точки (x2, y2): ').split(',')) 
distance = (((x2 - x1)**2 + (y2 - y1)**2)) ** 0.5
print(f'Расстояние между точками: {distance}')
```
