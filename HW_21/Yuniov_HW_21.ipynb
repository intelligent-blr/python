```
1. Напишите программу, которая подсчитывает количество вхождений каждого слова
в тексте и выводит на экран наиболее часто встречающиеся слова. Для решения
задачи используйте класс Counter из модуля collections. Создайте функцию
count_words, которая принимает текст в качестве аргумента и возвращает словарь
с количеством вхождений каждого слова. Выведите наиболее часто встречающиеся
слова и их количество.

Пример вывода:
Введенный текст: Lorem ipsum dolor sit amet, consectetur adipiscing elit. Sed sed lacinia est.

sed: 2
Lorem: 1
```

````python
from collections import Counter


def count_words(user_list):
    user_list = list(map(lambda word: word.replace(',', '').replace('.', ''), user_list))
    counter = Counter(user_list)
    most_common_word, count = counter.most_common(1)[0]
    return most_common_word, count

user_list = input("Введенный текст: ").lower().split()
most_common_word, count = count_words(user_list)

print(f"Самое часто встречающееся слово: '{most_common_word}', встречается {count} раз")
````

```
2. Напишите программу, которая создает именованный кортеж Person для хранения
информации о человеке, включающий поля name, age и city. Создайте список
объектов Person и выведите информацию о каждом человеке на экран.

Пример вывода:

Name: Alice, Age: 25, City: New York
Name: Bob, Age: 30, City: London
Name: Carol, Age: 35, City: Paris     
```

````python
from collections import namedtuple

Person = namedtuple('Person', ['Name', 'Age', 'City'])

info_people = [
    Person(Name = 'Alice', Age = 25, City = 'New York'),
    Person(Name = 'Bob', Age = 30, City = 'London'),
    Person(Name = 'Carol', Age = 35, City = 'Paris')
]

for person in info_people:
    print(f"Name: {person.Name}, Age: {person.Age}, City: {person.City}")
````

```
3. Напишите программу, которая принимает словарь от пользователя и ключ, и
возвращает значение для указанного ключа с использованием метода get или
setdefault. Создайте функцию get_value_from_dict, которая принимает словарь и
ключ в качестве аргументов, и возвращает значение для указанного ключа,
используя метод get или setdefault в зависимости от выбранного варианта.
Выведите полученное значение на экран.

Пример словаря:
my_dict = {'apple': 5, 'banana': 6, 'cherry': 7}

Пример вывода:

Введите ключ для поиска: banana
Использовать метод get (y/n)? y
Значение для ключа 'banana': 6


```

````python
def get_value_from_dict(my_dict):
    search_key = input('Введите ключ для поиска: ').lower()
    select_method = input('Использовать метод get (y/n)? ').lower()

    if select_method == 'y':
        value = my_dict.get(search_key, "Ключ не найден")
        print(f'Значение для ключа "{search_key}": {value}')
    elif select_method == 'n':
        value = my_dict.setdefault(search_key, 'Ключ не найден')
        print(f'Значение для ключа "{search_key}": {value}')
    else:
        raise ValueError('Неправильный выбор')

my_dict = {'apple': 5, 'banana': 6, 'cherry': 7}
get_value_from_dict(my_dict)
````
