```
1. Напишите программу, которая запрашивает у пользователя строку и определяет,
является ли она панграммой. Панграмма - это фраза, содержащая все буквы
алфавита. Программа должна игнорировать регистр букв и пробелы при проверке
панграммы. Выведите соответствующее сообщение на экран с помощью команды print.
Решить задачу для латиницы.

Пример вывода:
Введите строку: The quick brown fox jumps over the lazy dog
Строка является панграммой.
```

````python
a = input('Введите строку: ')
a = a.lower().replace(' ', '')
vse_bykvi = 'abcdefghijklmnopqrstuvwxyz'
sovpadenie = []
for i in vse_bykvi:
    if i in a and i not in sovpadenie:
        sovpadenie.append(i)
if len(sovpadenie) == len(vse_bykvi):
    print('Строка является панграммой')
else:
    print('Строка не является панграммой')
````

```
2. Напишите программу, которая запрашивает у пользователя строку и
выводит на экран количество гласных и согласных букв в ней.
Используйте функцию len() для подсчета количества букв.
Выведите результат на экран с помощью команды print.
Решить задачу для латиницы.

Пример вывода:
Введите строку: Hello World
Количество гласных букв: 3
Количество согласных букв: 7
```

````python
a = input('Введите строку: ')
a = a.lower().replace(' ', '')
glasnie = 'aeiouy'
soglasnie = 'bcdfghjklmnpqrstvwxz'
i_glasnie = []
i_soglasnie = []
for i in a:
    if i in glasnie:
        i_glasnie.append(i)
    elif i in soglasnie:
        i_soglasnie.append(i)
print(f'Количество согласных букв: {len(i_soglasnie)}')
print(f'Количество гласных букв: {len(i_glasnie)}')
````
