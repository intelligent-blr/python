```
1. Напишите программу, которая запрашивает у пользователя его имя,
возраст и место проживания, а затем выводит их в следующем формате:

"Привет, меня зовут {0}. Мне {1} лет. Я живу в {2}."

Вместо {0}, {1} и {2} подставьте соответствующие значения. Используйте метод
format() для форматирования строки. Потом попробуйте использовать f-строку.
Выведите результат на экран с помощью команды print.

Пример вывода:
Введите ваше имя: Alice
Введите ваш возраст: 25
Введите ваше место проживания: London
Привет, меня зовут Alice. Мне 25 лет. Я живу в London.
```

````python
name = input('Введите ваше имя: ')
age = input('Введите ваш возраст: ')
place_of_residence = input('Введите ваше место проживания: ')

# используя индексы
print('Привет, меня зовут {0}. Мне {1} лет. Я живу в {2}.'
.format(name, age, place_of_residence))

# используя ключи
print('Привет, меня зовут {имя}. Мне {возраст} лет. Я живу в {город}.'
.format(имя = name, возраст = age, город = place_of_residence))

# используя f-строку
print(f'Привет, меня зовут {name}. Мне {age} лет. Я живу в {place_of_residence}.')
````

```
2. Напишите программу, которая запрашивает у пользователя два числа и выводит
их сумму и произведение в следующем формате:

"Сумма: {sum:.2f}, Произведение: {product:.2f}"

Вместо {sum:.2f} и {product:.2f} подставьте соответствующие значения,
округленные до двух десятичных знаков. Используйте f-строки с использованием
форматных спецификаторов для форматирования чисел. Выведите результат на экран
с помощью команды print.

Пример вывода:
Введите первое число: 3.14159
Введите второе число: 2.71828
Сумма: 5.86, Произведение: 8.54
```

````python
n1 = float(input('Введите первое число: '))
n2 = float(input('Введите второе число: '))

сумма = n1 + n2
произведение = n1 * n2

print('Сумма: {sum:.2f}, Произведение: {product:.2f}'
.format(sum = сумма, product = произведение))

print(f'Сумма: {сумма:.2f}, Произведение: {произведение:.2f}')

print(f'Сумма: {n1 + n2:.2f}, Произведение: {n1 * n2:.2f}')
````
