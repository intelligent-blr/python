```
1. Напишите функцию, которая принимает список кортежей
от пользователя, где каждый кортеж содержит информацию
о студенте (имя, возраст, средний балл). Программа
должна вывести на экран имена студентов, чей средний
балл выше заданного значения. Используйте методы
кортежей и циклы для обработки данных. Выведите
итоговый список на экран с помощью команды print.

Пример списка кортежей:
[("Alice", 20, 90), ("Bob", 19, 80), ("Charlie", 21, 95), ("David", 18, 85)]

Пример вывода:
Введите пороговое значение среднего балла: 90
Студенты с средним баллом выше 90 : ['Charlie']
```

````python
def student_high_score():
    number_of_student = int(input('Количество студентов для анализа: '))
    list_student = []

    for i in range(number_of_student):
        student_name = input('Введите имя студента: ')
        student_age = int(input('Введите возраст возраст: '))
        student_score = int(input('Введите средний балл студента: '))
        student_inf = (student_name, student_age, student_score)
        list_student.append(student_inf)
    print(list_student)

    value = int(input('Введите пороговое значение среднего балла: '))
    names_students = []

    for student in list_student:
        if student[2] > value:
            names_students.append(student[0])
    print(f'Студенты с средним баллом выше 90 : {names_students}')
````

````python
student_high_score()
````

```
2. Напишите программу, которая принимает строку от
пользователя и разбивает ее на отдельные слова. Затем
программа должна создать новый кортеж, содержащий
длину каждого слова в исходной строке. Используйте
методы строк и кортежей для обработки данных.

Пример вывода:

Введите предложение: Программирование это интересно и полезно
Длины слов в предложении: (15, 3, 8, 2, 6)
```

````python
sentence = input('Введите предложение: ')
sentence_split = sentence.split()

len_words = []
for word in sentence_split:
    len_words.append(len(word))

result = tuple(len_words)
print(f'Длины слов в предложении: {result}')
````
