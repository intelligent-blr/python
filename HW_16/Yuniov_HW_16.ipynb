```
1. Напишите программу, которая принимает матрицу (вложенный список) от
пользователя и находит сумму всех элементов в матрице. Используйте вложенные
списки и циклы для обхода элементов матрицы.

Пример матрицы: [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

Пример вывода:
Сумма элементов в матрице: 45
```

````python
def sum_numbers_matrix():
    matrix = []

    while True:
        row_input = input("Введите числа, через запятую (или 'End' для завершения ввода): ")
        if row_input == 'End':
            break
        else:
            row = list(map(int, row_input.split(',')))
            matrix.append(row)

    total_sum = 0
    for row in matrix:
        for number in row:
            total_sum += number
    print(f'Сумма элементов в матрице: {total_sum}')
````

````python
sum_numbers_matrix()
````

```
2. Напишите программу, которая принимает список чисел от пользователя и
сортирует его в порядке убывания, используя метод sort() и параметр
reverse=True. Выведите отсортированный список на экран.

Пример вывода:

Введите список чисел, разделенных пробелами: 5 2 8 1 3
Отсортированный список чисел: [8, 5, 3, 2, 1]
```

````python
def list_numbers():
    list_number = list(map(int, input('Введите список чисел, разделенных пробелами: ').split()))
    print(list_number)
    list_number.sort(reverse=True)
    print(f'Отсортированный список чисел: {list_number}')
````

````python
list_numbers()
````
